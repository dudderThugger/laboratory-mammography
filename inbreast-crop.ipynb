```python
import numpy as np
import pandas as pd
from PIL import Image
from pathlib import Path 
import os
import matplotlib.pyplot as plt
import random

inbreast_path = '/media/szelesteya/F824D4D024D492CC/INBReast/inbreast-deprocessed'

crop_size = [224,224]
```

```python
image_names = []
anon = []
side = []
view_pos = []
acr = []
birads = []

for image_name in os.listdir(f"{inbreast_path}/images"):
    image_names.append(image_name)

    meta = image_name[0:-4].split('_')
    
    anon.append(meta[0])
    side.append(meta[1])
    view_pos.append(meta[2])
    acr.append(meta[4])
    birads.append(meta[6])

dict_imgs = {
    'anon': anon,
    'side': side,
    'view_pos': view_pos,
    'acr': acr,
    'birads': birads,
    'file': image_names
}

df_images = pd.DataFrame(dict_imgs)

df_images
```

```python
mask_names = []
anons = []
finding_types = []
coords_str = []
files = []

for image_name in os.listdir(f"{inbreast_path}/masks"):
    mask_names.append(image_name)
    meta = image_name[0:-4].split('_')
    anons.append(meta[0])
    if '.png' in image_name:
        finding_types.append(meta[1])
        coords_str.append(str([]))
    elif '.csv' in image_name:
        coords = np.genfromtxt(f"{inbreast_path}/masks/{image_name}",delimiter=',')
        finding_types.append('-'.join(meta[1:3]))
        coords_str.append(str(coords))
        
dict_findings = {
    'anon': anons,
    'finding_type': finding_types,
    'coord': coords_str,
    'mask_file': mask_names
}

df_findings = pd.DataFrame(dict_findings)
df_findings
```

```python
df_images_and_findings = df_findings.merge(df_images,on=['anon'])

df_images_and_findings
```

```python
df_calcs = df_images_and_findings[df_images_and_findings['finding_type'].isin(['calc','calc-grains'])]

df_calcs
```

```python
def scale_to_pixels(relative_coords, resolution):
    ret = np.zeros([relative_coords.shape[0], relative_coords.shape[1]])

    ret[0] = int(relative_coords[0] * resolution[1])
    ret[1] = int(relative_coords[1] * resolution[0])

    return ret

for i,calc_finding in df_calcs.loc[9:9,:].iterrows():
    image = Image.open(f"{inbreast_path}/images/{calc_finding['file']}")
    image_array = np.array(image)

    if calc_finding['finding_type'] == 'calc':
        mask = Image.open(f"{inbreast_path}/masks/{calc_finding['mask_file']}")
        mask_array = np.array(mask)

        masked_columns = np.nonzero(np.sum(mask_array, axis=0))
        x_min = masked_columns[0][0]
        x_max = masked_columns[0][-1]

        masked_rows = np.nonzero(np.sum(mask_array, axis=1))
        y_min = masked_rows[0][0]
        y_max = masked_rows[0][-1]

        mask_width = x_max - x_min
        mask_height = y_max - y_min

        # Choose random offset so the calcifications can be placed anywhere on the image not just the middle
        x_offset = random.randint(-crop_size[0] + mask_width, 0)
        y_offset = random.randint(-crop_size[1] + mask_height, 0)

        x = x_min + x_offset
        y = y_min + y_offset

        plt.imshow(image_array[y:y + croped_size[1],x:x + crop_size[0]], cmap='gray')
        plt.axis('off')
        plt.show()
    
    elif calc_finding['finding_type'] == 'calc-grains':
        grains_relative = np.genfromtxt(f"{inbreast_path}/masks/{calc_finding['mask_file']}",delimiter=',')
        size = image_array.shape

        # Change relative values to pixels
        grains_scaled = np.apply_along_axis(lambda x: scale_to_pixels(x, size), 0, grains_relative)
        print(grains_scaled)

        # x_min = np.min(grains_pixels[:,0])
        # x_max = np.max(grains_pixels[:,0])

        # y_min = np.min(grains_pixels[:,1])
        # y_max = np.max(grains_pixels[:,1])

        # plt.imshow(image_array[y_min:y_max,x_min:x_max], cmap='gray')
        # plt.axis('off')
        # plt.show()

```
